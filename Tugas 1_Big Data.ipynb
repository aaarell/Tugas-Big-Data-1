# Data Analysis

```python
merged_data = pd.merge(studentInfo, studentAssessment, on = 'id_student', how = 'left')
```

```python
# Mengganti nilai null dengan menghitung mean
mean_score = studentAssessment['score'].mean()
merged_data['score'].fillna(mean_score, inplace = True)
```

```python
# Mengelompokkan hasil akhir berdasarkan mean (nilai rata-rata)
grouped_data1 = merged_data.groupby('final_result')['score'].mean()
grouped_data1.head()
```

```python
# Mengelompokkan hasil akhir berdasarkan median (nilai tengah)
grouped_data2 = merged_data.groupby('final_result')['score'].median()
grouped_data2.head()
```

```python
# Mengelompokkan hasil akhir berdasarkan count (jumlah)
grouped_data3 = merged_data.groupby('final_result')['score'].count()
grouped_data3.head()
```

```python
# Hitung mean, median, dan jumlah skor berdasarkan hasil akhir
summary_group = merged_data.groupby('final_result')['score'].agg(['mean', 'median', 'count'])
summary_group.head()
```

## Visualisasi Dalam Bentuk Distribution

```python
# Mean (Nilai Rata-Rata)
grouped_data1.plot(kind = 'bar')
plt.xlabel('Final Result')
plt.ylabel('Score')
plt.title('Average Distribution Of Score By Final Result')
plt.show()
```

```python
# Median (Nilai Tengah)
grouped_data2.plot(kind = 'bar')
plt.xlabel('Final Result')
plt.ylabel('Score')
plt.title('Median Of Score By Final Result')
plt.show()
```

```python
# Count (Jumlah)
grouped_data3.plot(kind = 'bar')
plt.xlabel('Final Result')
plt.ylabel('Score')
plt.title('Number Of Student Score By Final Result')
plt.show()
```

## Visualisasi Dalam Bentuk Pie Chart

```python
# Mean (Nilai Rata-Rata)
fig = px.pie(names = grouped_data1.index, values = grouped_data1.values,
             title = 'Percentage of Mean Score By Final Result', width = 500, height = 500)
fig.update_traces(textposition = 'inside', textinfo = 'label+percent')
fig.show()
```

```python
# Median (Nilai Tengah)
fig = px.pie(names = grouped_data2.index, values = grouped_data2.values,
             title = 'Percentage of Median Score By Final Result', width = 500, height = 500)
fig.update_traces(textposition = 'inside', textinfo = 'label+percent')
fig.show()
```

```python
# Count (Jumlah)
fig = px.pie(names = grouped_data3.index, values = grouped_data3.values,
             title = 'Percentage Number Of Student Score By Final Result', width = 500, height = 500)
fig.update_traces(textposition = 'inside', textinfo = 'label+percent')
fig.show()
```
